# Instructions
Make a rock, paper, scissors game.

Inside the <code>main.py</code> file, you'll find the ASCII art for the hand signals already saved to a corresponding variable: <code>rock</code>, <code>paper</code>, and <code>scissors</code>. This will make it easy to print them out to the console.

Start the game by asking the player:

"What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors."

From there you will need to figure out:
- How you will store the user's input.
- How you will generate a random choice for the computer.
- How you will compare the user's and the computer's choice to determine the winner (or a draw).
- And also how you will give feedback to the player.

```python
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

#Write your code below this line ðŸ‘‡
import random 
print("Welcome to the rock-paper-scissors game")
choice = input("Please choose rock, paper, or scissors")
print(f"You choose {choice}")
lower_choice = choice.lower()
if(lower_choice == "rock"):
    print(rock)
elif(lower_choice == "paper"):
    print(paper)
else:
    print(scissors)

randomizer = random.randint(1, 3)
if(randomizer == 1):
    computer_choice = "rock"
    print(f"Computers choose {computer_choice}")
    print(rock)
    if(choice == "rock"):
        print("======It's a draw!======")
    elif(choice == "paper"):
        print("======You win!======")
    else:
        print("======You lose======")
elif(randomizer == 2):
    computer_choice = "paper"
    print(f"Computers choose {computer_choice}")
    print(paper)
    if(choice == "paper"):
        print("======It's a draw!======")
    elif(choice == "scissors"):
        print("======You win!======")
    else:
        print("======You loose======")
else:
    computer_choice = "scissors"
    print(f"Computers choose {computer_choice}")
    print(scissors)
    if(choice == "scissors"):
        print("======It's a draw!======")
    elif(choice == "rock"):
        print("======You win!======")
    else:
        print("======You loose======")
    
```
